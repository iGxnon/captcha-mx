```python
import PIL.Image as Image
import random
import os
import shutil
```

```python
def split(root, raw_dir, train_dir, test_dir, train_ratio=0.7, suffix='jpg'):
    raw_path = os.path.join(root, raw_dir)
    train_path = os.path.join(root, train_dir)
    test_path = os.path.join(root, test_dir)
    print(f'Start splitting dataset images of {raw_path} into {train_path} and {test_path}')
    print(f'Splitting at a ratio of {train_ratio*100}% train data')
    img_list = os.listdir(raw_path)
    img_list = [img for img in img_list if img.endswith(f'.{suffix}')]
    print(f'Total count of data is {len(img_list)}')
    train_list = random.choices(img_list, k=int(train_ratio*len(img_list)))
    tmp = set(train_list) # speed up searching
    test_list = [test_img for test_img in img_list if test_img not in tmp]
    print(train_list)
    del tmp
    print(f'Number of train data and test data is {len(train_list)}, {len(test_list)}')
    if not os.path.exists(train_path):
        os.mkdir(train_path)
    if not os.path.exists(test_path):
        os.mkdir(test_path)

    def move(_fnames, to):
        for fname in _fnames:
            src = os.path.join(raw_path, fname)
            dst = os.path.join(to, fname)
            shutil.move(src, dst)

    move(train_list, train_path)
    move(test_list, test_path)
    print('migration done')
```

```python
split(root=os.fspath('../sample'), raw_dir='train', train_dir='train', test_dir='valid')
```
